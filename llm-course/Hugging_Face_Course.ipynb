```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting for a meaning for the life.")
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been looking for a meaning for the life.")
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "I've been looking for a meaning for the life.",
    candidate_labels=["mental health", "Philosophy", "World", "animals"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("I've been looking for a meaning for the life.")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
generator(
    "I've been looking for a meaning for the life.",
    max_length=30,
    num_return_sequences=2,
)
```
